```python
def integer(x):
  t='0123456789'
  y=''
  for i in str(x):
    if i in t:
      y=y+i
  if y!='':
    return int(y)
  else:
    return 0
```

```python
Data['DomesticBox Office'] = Data['DomesticBox Office'].apply(integer)
```

```python
Data['WorldwideBox Office'] = Data['WorldwideBox Office'].apply(integer)
```

```python
Data['InternationalBox Office'] = Data['InternationalBox Office'].apply(integer)
```

```python
Data
```

```python
Data.isnull().sum()
```

```python
Data['DomesticBox Office'] = Data['DomesticBox Office'].replace(0, Data['DomesticBox Office'].mean())
```

```python
Data
```

```python
Data.duplicated().sum()
```

```python
pd.DataFrame(Data)
```

```python
Data_New = pd.DataFrame(Data)[['Rank', 'WorldwideBox Office', 'DomesticBox Office', 'InternationalBox Office']]
```

```python
Data_New
```

```python
from sklearn.model_selection import train_test_split
X = Data_New[['Rank','WorldwideBox Office','DomesticBox Office']]
Y = Data_New['InternationalBox Office']
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
Model = LinearRegression()
Model.fit(X_train,y_train)
```

```python
predictions = Model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, predictions)
print(f"Mean Squared Error : {mse}")
print(f"R2_Score : {r2_score(y_test,predictions)}")
```

###Mean Squared error is really high.Lets try to fix it by scaling.

```python
from sklearn.preprocessing import StandardScaler
X_train,X_test= StandardScaler().fit_transform(X_train),StandardScaler().fit_transform(X_test)
```

```python
y_train,y_test = StandardScaler().fit_transform(y_train.values.reshape(-1, 1)),StandardScaler().fit_transform(y_test.values.reshape(-1, 1))
```

```python
Model=LinearRegression()
Model.fit(X_train,y_train)
```

```python
predictions = Model.predict(X_test)
```

```python
print(f"Mean Squared Error : {mean_squared_error(y_test, predictions)}")
print(f"R2_Score : {r2_score(y_test,predictions)}")
```
